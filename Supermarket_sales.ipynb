```python
# import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Load the CSV into a DataFrame and take a peek at the first few rows.


```python
# Load the dataset
df = pd.read_csv("C:/Users/anjul/Downloads/SuperMarket_Analysis.csv")
```

- Check dataset info, descriptive statistics

```python
# Basic exploration
print("Dataset Info:")
print(df.info())
```

```python
print("\nDescriptive Statistics:")
print(df.describe())
```

Verify that there are no nulls in the dataset.

## üìä 5. Univariate Analysis

### 5.1 Distribution of Sales  
Use a histogram to understand the spread of the `Sales` column.


```python
# Histogram of Sales
sns.histplot(df['Sales'], bins=20, kde=True)
plt.title("Distribution of Sales")
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.show()
```

### 5.2 Sales by Gender  
Use a boxplot to compare sales distributions between genders.


```python
# Boxplot: Sales by Gender
sns.boxplot(data=df, x ='Gender', y = 'Sales')
plt.title("Sales by Gender")
plt.show()
```

### 5.3 Product Line Frequency  
Count how many transactions occurred in each product line.


```python
# Countplot: Product Lines
sns.countplot(data=df, y='Product line', order=df['Product line'].value_counts().index)
plt.title("Product Line Frequency")
plt.show()
```

## üîó 6. Bivariate Analysis

### 6.1 Quantity vs Sales  
Scatterplot to see how quantity sold relates to total sales.


```python
# Scatterplot: Quantity vs Sales
sns.scatterplot(data=df, x='Quantity', y='Sales')
plt.title("Quantity vs Sales")
plt.show()
```

### 6.2 Sales by City  
Boxplot to compare sales distributions across cities.


```python
# boxplot: City vs Sales
sns.boxplot(data=df, x='City', y='Sales')
plt.title("Sales Distribution by City")
plt.show()
```

### 6.3 Payment Method Distribution  
Countplot for payment types.


```python
# Countplot: Payment Method
sns.countplot(data=df, x='Payment')
plt.title("Payment Method Distribution")
plt.show()
```

## üîç 7. Correlation Analysis

Use a heatmap to uncover relationships among numeric features.


```python
# Correlation Heatmap
plt.figure(figsize=(8,4))
sns.heatmap(df.select_dtypes(include='number').corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

## üìà 9. Pairwise Relationships 

Use pairplot to visualize pairwise scatterplots and distributions.


```python
# pairplot 
sns.pairplot(df[['Sales', 'Quantity', 'Tax 5%', 'Unit price','cogs']])
plt.show()
```
